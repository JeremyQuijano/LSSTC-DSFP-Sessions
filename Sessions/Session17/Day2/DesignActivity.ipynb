```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
data_df = pd.read_csv('AIR_GHG_28022023214516359.csv')
data_df
```

```python
data_grouped = data_df.groupby(['Country'])['Value'].sum()
print(data_grouped)
# data_grouped['Year']
np.sort(data_grouped, axis=-1)
```

```python
%%time
groups = data_df.groupby('Country')
plt.figure(figsize=(10, 5))
for name, group in tqdm(groups):
    plt.bar(group['Year'], group['Value'])

plt.title('Total Greenhouse Gas Emission (Linear)')
plt.xlabel('Year')
plt.ylabel('Tonnes')
plt.show()

# for name, group in tqdm(groups):
#     plt.bar(group['Year'], group['Value'])

# plt.title('Total Greenhouse Gas Emission (Log)')
# plt.yscale('log')
# plt.xlabel('Year')
# plt.ylabel('Tonnes')
# plt.show()
```
